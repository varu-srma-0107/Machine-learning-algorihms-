```python
import pandas as pd
from sklearn.datasets import load_iris
iris = load_iris()
dir (iris)
```

```python
df = pd.DataFrame(iris.data,columns=iris.feature_names)
df.head()
```

```python
df['target']= iris.target
```

```python
df[df.target==2].head()
```

```python
df['flower_name']= df.target.apply(lambda x: iris.target_names [x])
```
