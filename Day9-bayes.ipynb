```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\LENOVO\Desktop\VS Code\ML Projects\Code_basics-ML-Tutorial\titanic.csv")
```

```python
df.head()
```

```python
target = df['Survived']
df.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'],axis='columns',inplace=True)

input = df.drop('Survived',axis ='columns')
dummies = pd.get_dummies(input.Sex)
```

```python
dummies.head()
```

```python
input = pd.concat([input,dummies],axis = 'columns')
```

```python
input.head()
```

```python
input.columns[input.isna().any()]
```

```python
input.Age[:10]
```

```python
input.Age = input.Age.fillna(input.Age.mean())
input.head()
```

```python
input.drop(['Sex','male'],axis='columns',inplace=True)
input.head(3)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(input,target,test_size=0.3)
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
```

```python
model.fit(X_train,y_train)
```

```python
X_test[0:10]
```

```python
y_test[0:10]
```

```python
model.predict(X_test[0:10])
```

```python
model.predict_proba(X_test[:10])
```

```python
model.score(X_test,y_test)
```
